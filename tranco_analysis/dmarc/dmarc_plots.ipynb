```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
matplotlib.rcParams["pdf.fonttype"] = 42
matplotlib.rcParams["ps.fonttype"] = 42
```

```python

# Data
tiers = ["Tier 1 (1-1K)", "Tier 2 (1K-10K)", "Tier 3 (10K-100K)", "Tier 4 (100K-1M)"]
present = [735, 5820, 50237, 390810]
valid = [728, 5765, 49670, 385779]

x = np.arange(len(tiers))
width = 0.35

# Plot
plt.figure()
plt.bar(x - width/2, present, width, label='Present')
plt.bar(x + width/2, valid, width, label='Valid')

# Labels and formatting
plt.xlabel("Ranking Tier")
plt.ylabel("Count")
plt.title("DMARC Adoption by Ranking Tier")
plt.xticks(x, tiers, rotation=20)
plt.legend()

plt.tight_layout()
# plt.savefig("dmarc_plots/dmarc_adoption_rate_per_ranking.pdf")
plt.show()
```

```python
# Data
tiers = ["Tier 1 (1-1K)", "Tier 2 (1K-10K)", "Tier 3 (10K-100K)", "Tier 4 (100K-1M)"]
present = [735, 5820, 50237, 390810]
valid = [728, 5765, 49670, 385779]
totals = [1000, 9000, 90000, 900000]

# Convert to percentages
present_pct = [(p / t) * 100 for p, t in zip(present, totals)]
valid_pct = [(v / t) * 100 for v, t in zip(valid, totals)]

x = np.arange(len(tiers))
width = 0.35

# Plot
plt.figure()
plt.bar(x - width/2, present_pct, width, label='Present (%)')
plt.bar(x + width/2, valid_pct, width, label='Valid (%)')

# Labels and formatting
plt.xlabel("Ranking Tier")
plt.ylabel("Percentage (%)")
plt.ylim([0,100])
plt.title("DMARC Adoption by Ranking Tier (Percentage)")
plt.xticks(x, tiers, rotation=20)
plt.legend()
plt.tight_layout()

plt.savefig("dmarc_plots/dmarc_adoption_rate_per_ranking.pdf")

plt.show()
```

```python
# Data
labels = [
    "none",
    "quarantine",
    "reject",
    "missing"
]

sizes = [228627, 106668, 106647, 5660]

# Plot
plt.figure()
wedges, texts, autotexts = plt.pie(
    sizes,
    autopct='%1.2f%%',
    startangle=90
)

# Add legend outside
plt.legend(
    wedges,
    labels,
    title="Policy",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.title("DMARC Policy Distribution (p=)")
plt.tight_layout()
plt.axis('equal')  # Ensures the pie is a circle
plt.savefig("dmarc_plots/policy_distribution.pdf", bbox_inches='tight')

plt.show()
```
