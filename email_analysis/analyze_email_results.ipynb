```python
import pandas as pd
```

```python
df1 = pd.read_csv("elizabeth_email_auth_results.csv")
df2 = pd.read_csv("shreyashi_email_auth_results.csv")
df3 = pd.read_csv("veena_email_auth_results.csv")

df = pd.concat((df1, df2, df3), axis=0)
df.columns = ["domain","spf_result","dkim_result","dmarc_result","dmarc_policy"]
print(df.shape)
```

# All pass rate

```python
all_pass_rate = (len(df[(df["spf_result"] == "pass") & (df["dkim_result"] == "pass") & (df["dmarc_result"] == "pass")]) / len(df)) * 100
print(all_pass_rate)
```

# SPF pass rate

```python
spf_rate = (len(df[(df["spf_result"] == "pass")]) / len(df)) * 100
print(spf_rate)
```

# DKIM pass rate

```python
dkim_rate = (len(df[(df["dkim_result"] == "pass")]) / len(df)) * 100
print(dkim_rate)
```

# DMARC pass rate

```python
dmarc_rate = (len(df[(df["dmarc_result"] == "pass")]) / len(df)) * 100
print(dmarc_rate)
```

## DMARC policy

```python
dmarc_policy_none = (len(df[(df["dmarc_policy"] == "NONE")]) / len(df)) * 100
print(dmarc_policy_none)
```

```python
dmarc_policy_quarantine = (len(df[(df["dmarc_policy"] == "QUARANTINE")]) / len(df)) * 100
print(dmarc_policy_quarantine)
```

```python
dmarc_policy_reject = (len(df[(df["dmarc_policy"] == "REJECT")]) / len(df)) * 100
print(dmarc_policy_reject)
```
