```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
matplotlib.rcParams["pdf.fonttype"] = 42
matplotlib.rcParams["ps.fonttype"] = 42
```

```python


# Data
categories = ["Promotions", "Social", "Updates"]

spf = [100.00, 100.00, 100.00]
dkim = [100.00, 100.00, 95.83]
dmarc = [94.74, 100.00, 95.83]
all3 = [94.74, 100.00, 91.67]

x = np.arange(len(categories))
width = 0.2

# Plot
plt.figure()

plt.bar(x - 1.5*width, spf, width, label='SPF')
plt.bar(x - 0.5*width, dkim, width, label='DKIM')
plt.bar(x + 0.5*width, dmarc, width, label='DMARC')
# plt.bar(x + 1.5*width, all3, width, label='All 3')

# Labels and formatting
plt.xlabel("Email Category")
plt.ylabel("Pass Rate (%)")
plt.title("Email Authentication Pass Rates by Category")
plt.xticks(x, categories)
plt.ylim(0, 110)

# Legend moved outside
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig("pass_rates")
plt.show()
```
